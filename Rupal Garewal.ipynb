## Missing values in Columns and rows

```python
#Calculate missing value %
df_null_pct = df.isnull().sum()/len(df)*100
df_null_pct.sort_values(ascending = False)
```

```python
 #Filtering columns with >= to 50% missing of values which will be our peek to drop the columns
df = df.loc[:,df_null_pct<50]
```

```python
df.shape
```

```python
df.dtypes
```

```python
# List of col which are not required 21 columns
ldrop_col1 = ['delinq_2yrs','earliest_cr_line','inq_last_6mths','open_acc','pub_rec','revol_bal','revol_util','total_acc','out_prncp','out_prncp_inv','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d','application_type']
# Drop the col nr 21 columns
df.drop(ldrop_col1,axis=1,inplace = True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.shape
```

```python
#List of the columns having one value
unique = df.nunique()
unique.sort_values()


#Whether any cols have one value
listcolunival = unique[unique.values==1]
print(listcolunival)

#Drop the columns that are having single value
ldrop_col2=listcolunival.index.tolist()
df.drop(ldrop_col2,axis=1,inplace = True)
df.shape
```

```python
#drop the columns which are not needed for analysis 
ldrop_col3 = ["id", "member_id", "url", "desc","zip_code"]
df.drop(ldrop_col3,axis=1,inplace = True)
df.shape
```

```python
 #Filter rows with nonnull values for pub_rec_bankruptcies
df = df[~df.pub_rec_bankruptcies.isnull()]
df.shape
#View  all data types
df.dtypes
```

### Changing objects to correct data type

- funded_amnt_inv    to       int64
- annual_inc         to       int64
- addr_state         to       object
- issue_d            to       Datetime
- term               to       int64 after removing month string
- int_rate           to       float64 
- replacements + and <






```python
#Convert "annual_inc" to int64
print(df.annual_inc.dtype)
df.annual_inc = df.annual_inc.astype("int64")
print(df.annual_inc.dtype)
```

```python
#Convert "term" to string
print(df.term.dtype)
df.term = df.term.astype("str")
print(df.term.dtype)
```

```python
#Convert "funded_amnt_inv" to int64
print(df.funded_amnt_inv.dtype)
df.funded_amnt_inv = df.funded_amnt_inv.astype("int64")
print(df.funded_amnt_inv.dtype)
```

```python
#emp val count

df.emp_length.value_counts()
```

```python
#Convert issue_d to Datetime
print(df.issue_d)
df.issue_d = pd.to_datetime(df.issue_d, format='%b-%y')
print(df.issue_d)
print(df.issue_d.dtype)
```

```python
df.dtypes
```

```python
#consider values "Fully Paid" and "Charged Off" and check the shape
df = df[df.loan_status.isin(["Fully Paid","Charged Off"])]
df.shape
```

```python
#replacements + and <
df.emp_length = df.emp_length.str.replace("+",'')
df.emp_length = df.emp_length.str.replace("<",'')
df.emp_length.value_counts()
```

```python
# Converting months to int64
print(df.term)
df.term = df.term.str.replace(" months",'').astype('int64')
print(df.term)
print(df.term.dtype)
```

```python
#Convert int_rate to float64 
print(df.int_rate)
#removing "%" symbol
df.int_rate = df.int_rate.str.replace("%",'').astype('float64')
print(df.int_rate)
```

## Derived Metrics
- Derived metrics for Month and Year from column "issue_d"
- Derived metrics for "loan_amnt" to "annual_inc" ratio
- Create new column loan_statC with 0 and 1 values based on loan_status column where 0="Fully Paid" and 1="Charged off"


### Derived columns - Lets derive some new columns using existing data as it may further used in analysis.

```python
#Derived metrics analysis on column "issue_d"

#create month and year columns separately

df.issue_d = pd.to_datetime(df.issue_d, format='%b-%y')
df['year']=df['issue_d'].dt.year
df['month']=df['issue_d'].dt.month
df.head()
```

```python
#Calculate Loan to Amount income ratio
df["amnt_to_inc_ratio"] = df.loan_amnt/df.annual_inc
df[["loan_amnt", "annual_inc","amnt_to_inc_ratio"]]
```

```python
#Create new column loan_statC with 0 and 1 values based on loan_status column where 0="Charged Off" and 1="Fully Paid"
loan_status_map = {'Charged Off':0,'Fully Paid':1}
df.loan_statC = df.loan_status.map(loan_status_map)
print(df.loan_statC)
```
